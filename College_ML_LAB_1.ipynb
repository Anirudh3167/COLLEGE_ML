```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Excel file from Google Drive
file_path = 'C:\Users\Anirudh\Downloads\Sample Stats data.xlsx'
df = pd.read_excel(file_path)

# Display the DataFrame
print("DataFrame:")
print(df)
```

```python
# Basic Data Analysis
# Sorting by 'Students Selected'
sorted_df = df.sort_values(by='Students Selected', ascending=False)
print("\nSorted DataFrame by Students Selected:")
print(sorted_df)

# Filtering companies with more than 10 students selected
filtered_df = df[df['Students Selected'] > 10]
print("\nFiltered DataFrame (Students Selected > 10):")
print(filtered_df)
```

```python
# Visualization
# Create a bar plot for Students Registered vs Students Selected
plt.figure(figsize=(10, 6))
sns.barplot(x='Company', y='Students Registered', data=df, color='blue', label='Registered')
sns.barplot(x='Company', y='Students Selected', data=df, color='orange', label='Selected')
plt.title('Students Registered vs Selected by Company')
plt.xlabel('Company')
plt.ylabel('Number of Students')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Install and import necessary libraries
!pip install ipywidgets
```

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a slider widget
slider = widgets.IntSlider(
    value=10,
    min=0,
    max=100,
    step=1,
    description='Value:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
```

```python
# Define a function to modify the plot based on the slider's input
def update_plot(change):
    value = change['new']
    plt.figure(figsize=(8, 6))
    x = np.linspace(0, 10, 100)
    y = np.sin(x * value)
    plt.plot(x, y)
    plt.title(f'Sine Wave with Frequency: {value}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Step 4: Link the slider to the update_plot function
slider.observe(update_plot, names='value')

# Step 5: Display the slider and plot
display(slider)
```
